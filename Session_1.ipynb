```python
#Tokenizing
from nltk import word_tokenize, sent_tokenize
text = "The sun rises in the east, and it sets in the west. Many people enjoy watching the sunrise over the mountains."
tokenized_text = word_tokenize(text)
tokenized_sentence = sent_tokenize(text)
print(tokenized_text)
print(tokenized_sentence)
```

```python
#Stopwords
from nltk.corpus import stopwords
eng_stopwords = stopwords.words('english')
# print(eng_stopwords)
removed_stopwords_list = [word for word in tokenized_text if word.lower() not in eng_stopwords]
for word in tokenized_text:
    if word.lower not in eng_stopwords:
        removed_stopwords_list.append(word)
print(removed_stopwords_list)
```
